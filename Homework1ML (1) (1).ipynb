**17\.** Как получить день месяца, номер недели, день года и день недели из даты в виде строки? (★★☆)

```python
s = pd.Series(['01 Jan 2010', '02-02-2011', '20120303', '2013/04/04', '2014-05-05', '2015-06-06T12:20'])
dates = pd.DatetimeIndex(s)
def get_date_attr(element_index):
    return dates[element_index]

print("Date: ", get_date_attr(0).date())

print("Week number: ", get_date_attr(1).week)

print("Day number of year: ", get_date_attr(2).year)

print("Day of week: ", get_date_attr(3).dayofweek)
```
